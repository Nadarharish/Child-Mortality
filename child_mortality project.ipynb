```python
plt.figure(figsize = (20,20))
plt.subplot(2,2,1)
plt.title("YEAR VS POPULATION (INDIA)")
plt.scatter(x='year', y='totalpopulation', data=india)
plt.subplot(2,2,2)
plt.title("YEAR VS POPULATION (CHINA)")
plt.scatter(x='year', y='totalpopulation', data=china)
plt.subplot(2,2,3)
plt.title("YEAR VS CHILDMORTALITY (INDIA)")
plt.scatter(x='year', y='childmortality', data=india)
plt.subplot(2,2,4)
plt.title("YEAR VS CHILDMORTALITY (CHINA)")
plt.scatter(x='year', y='childmortality', data=china)
```

#### The above subplot compares the population of India and China over years and the childmortality of the India and China over years. From the plot we can clearly identify the increasing trend in population of both the countries but the rate of increase in population in India is high as compared to China and the child mortality is exactly opposite to the trend of the population in india. in the initial years there is a slight increase in the child mortality then after 1972 there is a decrease trend in the child mortality. it shows an inverse relationship between india's population and child mortality. whereas on the other hand the childmortality in China is fluctuating. it does'nt shows any relationship. 

```python
year = mortality.groupby(['year']).sum('childmortality')
year.shape
```

```python
year['year']=year.index
```

```python
sns.catplot(x='totalpopulation',y = 'year',data=year.iloc[1:], kind='bar', height=40, aspect=1, orient='h')
```

```python
sns.catplot(x='mortalityrate',y = 'year',data=year.iloc[1:], kind='bar', height=40, aspect=1, orient='h')
```

```python
plt.figure(figsize = (20,5))
plt.subplot(1,2,1)
plt.title("YEAR VS POPULATION")
plt.scatter(x ='year',y ='totalpopulation',data=year)
plt.subplot(1,2,2)
plt.title("YEAR VS MORTALITY RATE")
plt.scatter(x='year', y ='mortalityrate',data = year)
```

#### The above subplot compares the total population and the mortality rate of 194 countries of 65 years. population shows an positive or increase trend and mortality rate shows an inverse relation to the population. population increases due to decreases in child mortality.
